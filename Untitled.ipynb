### Creating a rolling (moving) average of your data and visualize it.  We can calculate moving averages with the .rolling method.

```python
y.plot(alpha=.2, label='Daily')
y.resample('W').mean().rolling(3).mean().plot(alpha=.5, label='3 weeks')
y.resample('M').mean().rolling(7).mean().plot(alpha=.8, label='7 month')
plt.legend()
```

```python
y.plot(alpha=.2, label='Daily')
y.resample('M').mean().rolling(3).mean().plot(alpha=.5, label='3 Months')
y.resample('M').mean().rolling(6).mean().plot(alpha=.8, label='6 Months')
plt.legend()
```
